**A)** Para el caso $I_e=0$


$$
\frac{d V_m(t)}{dt}=\frac{1}{\tau_m} (E_L-V_m(t))
$$

$\frac{d V_m(t)}{dt}=f(V_m)=0 \hspace{0.5cm} \rightarrow\hspace{0.5cm} \frac{1}{\tau_m} (E_L-V_m(t))=0 \hspace{0.5cm} \rightarrow\hspace{0.5cm} V_m(t)=E_L$ 


$f(V_m)=\frac{1}{\tau_m} (E_L-V_m(t)) \hspace{0.5cm} \rightarrow\hspace{0.5cm}
\frac{df}{dV_m}=-\frac{1}{\tau} \\ \\ $


 $ \frac{df}{dV_m} < 0 $  para todo $V_m$, en especial para $V_m=E_L$, que es el punto fijo, por lo tanto es un punto fijo estable.




```python
n=50
tau=10
El=-65


x_punto=np.zeros(n)
x=np.linspace(El-5,-40,n)




for i in np.arange(0,n):
  x_punto[i]=1/tau*(El-x[i])


line_y0=np.zeros(n)

plt.plot(x,x_punto,color='black')
plt.scatter(El,0, color='black') #, facecolors='none'
plt.plot(x,line_y0,'k--')
plt.arrow(El-6,0,3,0,color='blue', length_includes_head=True,head_width=0.35, head_length=0.8)
plt.arrow(El+6,0,-3,0,color='blue', length_includes_head=True,head_width=0.35, head_length=0.8)
plt.arrow(El+12,0,-3,0,color='blue', length_includes_head=True,head_width=0.35, head_length=0.8)
plt.show()
```

**C)** Para el caso $I_e=I$

$$
\frac{d V_m(t)}{dt}=\frac{1}{\tau_m} (E_L-V_m(t)+R_m I)
$$
\
$$
\frac{dV_m(t)}{E_L-V_m(t)+R_mI)}=\frac{1}{\tau_m}dt
$$
\
$x=E_L-V_m(t)+R_mI \hspace{0.5cm} \rightarrow \hspace{0.5cm} dx=-dV_m(t)$
\
\
$$
-\frac{dx}{x}=\frac{1}{\tau_m}dt \hspace{0.5cm} \rightarrow \hspace{0.5cm} -ln(x)=\frac{t}{\tau_m}+V_0 \hspace{0.5cm} \rightarrow \hspace{0.5cm} x=e^{-t/\tau_m-V_0}
$$
\
\
Reemplazando $A=e^{-V_0}$ y $x=E_L-V_m(t)+R_mI$ se tiene
\
\
$$
E_L-V_m(t)+R_mI=A e^{-t/\tau_m}
$$
\
$$
\boxed{ V_m(t)=E_L+R_m I- A e^{-t/\tau_m} }
$$

para $t=0$ 
$$
V_m(0)=E_L+R_mI-A \hspace{0.5cm} \rightarrow \hspace{0.5cm} A=E_L+R_mI-V_m(0)
$$






```python
tau=10
El=-65
Rm=10



m=200

V0=El

t=np.linspace(0,200,m)

V_exacto=np.zeros(m)

colors=['blue','green','red','black', 'gray']

for Na in np.arange(-2,3):
  A=El+Rm*2*Na-V0
  for i in np.arange(0,m):
    V_exacto[i]=El+Rm*2*Na-A*np.exp(-t[i]/tau)

  plt.plot(t,V_exacto,color=colors[Na+2], label='N*a='+str(Na))
  

line_y0=np.zeros(m)
plt.plot(t,line_y0,'k--')
plt.legend(loc='upper left')
plt.show()
```

**E)** Método de Rung-Kutta para 

$$ 
\frac{dV_m(t)}{dt}=\frac{1}{\tau_m}(E_L-V_m(t)+R_mI_c(t)), \hspace{0.4cm} \rightarrow \hspace{0.4cm} V(t=0)=E_L \hspace{0.4cm} \rightarrow \hspace{0.4cm}  0ms \leq t \leq 200ms
$$
\
$h=0.05ms$ e $I_e=2nA$ 


```python
tau=10
El=-65
Rm=10

h=0.05
dt=h

m=200/h

V0=El

t=np.linspace(0,200,int(m))


colors=['blue','green','red','magenta', 'cyan']
colors_punteado=['b--','g--','r--','m--', 'c--']



def f(El,Rm,nA,tau, V,t):
  y=1/tau*(El-V+Rm*2*nA)
  return y

dt=t[1]-t[0]
for nA in np.arange(-2,3):
  V=np.zeros(int(m))
  V[0]=V0
  V_exacto=np.zeros(int(m))
  A=El+Rm*2*nA-V0
  for i in np.arange(0,int(m-1)):

    k1 = f(El,Rm,nA,tau,V[i],t[i])
    k2 = f(El,Rm,nA,tau,V[i]+0.5*h*k1,t[i]+0.5*dt)
    k3 = f(El,Rm,nA,tau,V[i]+0.5*h*k2,t[i]+0.5*dt)
    k4 = f(El,Rm,nA,tau,V[i]+h*k3,t[i]+dt)

    V[i+1]=V[i]+h*(k1+2.0*k2+2.0*k3+k4)/6.0

    V_exacto[i]=El+Rm*2*nA-A*np.exp(-t[i]/tau)


  plt.plot(t[0:int(m-1)],V_exacto[0:int(m-1)],color=colors[nA+2], label='N*a='+str(nA))
  plt.plot(t,V,colors_punteado[nA+2])#, label='N*a='+str(nA))

line_V0=np.zeros(int(m))
line_t0=np.zeros(int(m))
axe_V=np.linspace(-120,10,int(m))

plt.plot(t,line_V0,'black')
#plt.plot(line_t0,axe_V,'black')
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

**F)**

$$

$$
\
con los parámetros anteriores. Ahora utilizar un valor de potencial umbral $V_u=-50mV$

```python
tau=10
El=-65
Rm=10
I=2
Vu=-50

h=0.05
dt=h

m=200/h

V0=El

t=np.linspace(0,200,int(m))


def f(El,Rm,nA,tau,I, V,t):
  y=1/tau*(El-V+Rm*I)
  return y

dt=t[1]-t[0]

V=np.zeros(int(m))
V[0]=V0
V_exacto=np.zeros(int(m))
A=El+Rm*I-V0

t_periodo=[]
for i in np.arange(0,int(m-1)):

  k1 = f(El,Rm,nA,tau,I,V[i],t[i])
  k2 = f(El,Rm,nA,tau,I,V[i]+0.5*h*k1,t[i]+0.5*dt)
  k3 = f(El,Rm,nA,tau,I,V[i]+0.5*h*k2,t[i]+0.5*dt)
  k4 = f(El,Rm,nA,tau,I,V[i]+h*k3,t[i]+dt)

  V[i+1]=V[i]+h*(k1+2.0*k2+2.0*k3+k4)/6.0

  if V[i+1]>Vu:
    V[i+1]=V0
    t_periodo.append(t[i])

  V_exacto[i]=El+Rm*I-A*np.exp(-t[i]/tau)


plt.plot(t[0:int(m-1)],V_exacto[0:int(m-1)],'red', label='Exacta')
plt.plot(t,V,'b--', label='Rung-Kutta')

line_V0=np.zeros(int(m))
line_t0=np.zeros(int(m))
axe_V=np.linspace(-120,10,int(m))

#plt.plot(t,line_V0,'black')
#plt.plot(line_t0,axe_V,'black')
plt.legend(bbox_to_anchor=(1, 1))
plt.show()


periodo=[]
for i in np.arange(0,len(t_periodo)-1):
  periodo.append(t_periodo[i+1]- t_periodo[i])

print(periodo)
```

**H)** La frecuencia puede obtener con el tiempo en que el sistema tarda en alcanzar el valor umbral. Teniendo la solución
$$
V_m(t)=E_L+R_m I- A e^{-t/\tau_m} 
$$

con $ A=E_L+R_mI-V_m(0)$. Cuando $V_m(t)=V_u$ 

$$
V_u=E_L+R_m I- A e^{-t/\tau_m} 
$$
\
$$
V_u-E_L-R_mI=-A e^{-t/\tau_m}
$$
\
$$
\frac{-V_u+E_L+R_mI}{A}=e^{-t/\tau_m}
$$
\
$$
\frac{-V_u+E_L+R_mI}{E_L+R_mI-V_m(0)}=e^{-t/\tau_m}
$$
\
$$
t=-\tau_m ln \left(  \frac{-V_u+E_L+R_mI}{E_L+R_mI-V_m(0)} \right)
$$
\
Reemplazando $\tau_m=10ms$, $E_L=-65mV$, $R_m=10 M \Omega$, $V_u=-50mV$ y $V_m(0)=E_L$
\
$$
t=-10ms \hspace{0.15cm} ln \left(  \frac{50 mV-65mV+10 M \Omega \hspace{0.15cm}I}{10 M \Omega \hspace{0.15cm} I} \right)
$$
\
$$
t=-10ms \hspace{0.15cm} ln \left(  \frac{-15mV+10 M \Omega \hspace{0.15cm}I}{10 M \Omega \hspace{0.15cm} I} \right)
$$
\
$$
t=-10ms \hspace{0.15cm} ln \left(  \frac{-1.5nA}{ I}+1 \right)
$$
